```python
img_dic = {}

folder_path = '/mnt/d/download/'

# 배열 평탄화
def flatten(lst):
    flat_list = []
    for item in lst:
        if isinstance(item, list):
            flat_list.extend(flatten(item))
        else:
            flat_list.append(item)
    return flat_list

# 원본 이미지 파일
ori_img = glob(f'{folder_path}CelebAMask-HQ/CelebA-HQ-img/*')
# 마스크 이미지 파일
mask_img = os.listdir(f'{folder_path}CelebAMask-HQ/CelebAMask-HQ-mask-anno/')[1:]
# 모든 마스크 이미지 중에서 hair 마스크만 추출
hair_mask = []
for n in mask_img:
    mask = [i for i in glob(f'{folder_path}CelebAMask-HQ/CelebAMask-HQ-mask-anno/{n}/*') if 'hair' in i]
    hair_mask.append(mask)

# 평탄화
# hair_mask = flatten(hair_mask)

print(len(hair_mask))
```

```python
# os.path.basename(hair_mask[-1]).split('_')[0]
# os.path.dirname(hair_mask[-1])

for original_image in ori_img:
    # 파일 이름
    img_name = os.path.basename(original_image)
    # 파일 위치
    img_path = os.path.dirname(original_image)
    
    # 원본 이미지가 마스크 이미지에 있는지 확인 하고 있으면 디셔너리 만듬
    img_num = int(img_name[:-4]) // 2000
    print(img_num, img_name)
    for mask_img in hair_mask[img_num]:
        if img_name[:-4] in mask_img:
            img_dic[os.path.join(img_path, img_name)] = f'{folder_path}CelebAMask-HQ/CelebAMask-HQ-mask-anno/{img_num}/{img_name[:-4]}_hair.jpg'            
```
